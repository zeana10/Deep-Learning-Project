```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None) 
df = pd.read_csv("csv/dt_mrg.csv")
```

```python
df['gentrify_flag'].replace(['Gentrified', 'Gentrifying', 'Non-Gentrified'], [2,1,0],inplace=True)
df['Poverty_flag'].replace([False, True], [0,1],inplace=True)
```

```python
df.head()
```

```python
location = [i for i in range(len(sorted(list(set(df['Sub-Borough Area'])))))]
df['Sub-Borough Area'].replace(sorted(list(set(df['Sub-Borough Area']))),location,inplace=True)
df.head()
```

```python
from keras.utils import to_categorical
y_label = list(df['gentrify_flag'])
df.drop(['gentrify_flag'],axis=1,inplace=True)
df.drop(df.columns[[0]],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.shape
```

split the train and test set

```python
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation
from sklearn.model_selection import train_test_split
from keras.layers import Input, MaxPooling2D, UpSampling2D
from keras.models import Model
from keras.layers.convolutional import Conv2D

y_encoded = to_categorical(y_label)
x_train, x_test, y_train, y_test = train_test_split(df, y_encoded, test_size=0.2, random_state=0)
print(x_test.shape)
```

add auto encoder for x_train

```python
input_img = Input(shape=(24,))
encoded = Dense(units=12, activation='relu')(input_img)
encoded = Dense(units=6, activation='relu')(encoded)
encoded = Dense(units=3, activation='relu')(encoded)
decoded = Dense(units=6, activation='relu')(encoded)
decoded = Dense(units=12, activation='relu')(decoded)
decoded = Dense(units=24, activation='sigmoid')(decoded)
autoencoder=Model(input_img, decoded)
#encoder = Model(input_img, encoded)
autoencoder.summary()
#encoder.summary()
autoencoder.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
autoencoder.fit(x_train, x_train,
                epochs=50,
                batch_size=3,
                shuffle=True,
                validation_data=(x_test, x_test))
```

```python
model = Sequential()
model.add(Dense(48, input_dim=24, init='uniform', activation='relu'))
#model.add(Dropout(0.2))
model.add(Dense(24, init='uniform', activation='relu'))
#model.add(Dropout(0.2))
model.add(Dense(3, init='uniform', activation='softmax'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_train, y_train, nb_epoch=500, batch_size=50)
train_acc = model.evaluate(x_train, y_train)
print("%s: %.2f%%" % (model.metrics_names[1], train_acc[1]*100))
    
test_acc = model.evaluate(x_test, y_test)
print("%s: %.2f%%" % (model.metrics_names[1], test_acc[1]*100))
```

denoise auto encoder

```python
noise_factor = 0.5
x_train_noisy = x_train+ noise_factor * np.random.normal(loc=0.0, scale=1.0, size=x_train.shape)
x_test_noisy = x_test + noise_factor * np.random.normal(loc=0.0, scale=1.0, size=x_test.shape)
x_train_noisy = np.clip(x_train_noisy, 0., 1.)
x_test_noisy = np.clip(x_test_noisy, 0., 1.)
```

```python
input_img = Input(shape=(24,))
encoded = Dense(units=12, activation='relu')(input_img)
encoded = Dense(units=6, activation='relu')(encoded)
encoded = Dense(units=3, activation='relu')(encoded)
decoded = Dense(units=6, activation='relu')(encoded)
decoded = Dense(units=12, activation='relu')(decoded)
decoded = Dense(units=24, activation='sigmoid')(decoded)
autoencoder=Model(input_img, decoded)
#encoder = Model(input_img, encoded)
autoencoder.summary()
#encoder.summary()
autoencoder.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
autoencoder.fit(x_train_noisy, x_train_noisy,
                epochs=50,
                batch_size=3,
                shuffle=True,
                validation_data=(x_test_noisy, x_test_noisy))
```

```python
model = Sequential()
model.add(Dense(48, input_dim=24, init='uniform', activation='relu'))
#model.add(Dropout(0.2))
model.add(Dense(24, init='uniform', activation='relu'))
#model.add(Dropout(0.2))
model.add(Dense(3, init='uniform', activation='softmax'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_train, y_train, nb_epoch=500, batch_size=50)
train_acc = model.evaluate(x_train_noisy, y_train)
print("%s: %.2f%%" % (model.metrics_names[1], train_acc[1]*100))
    
test_acc = model.evaluate(x_test, y_test)
print("%s: %.2f%%" % (model.metrics_names[1], test_acc[1]*100))
```
